```python
import subprocess
import sys
import os
import platform
import shutil



# Función para verificar si Docker está instalado

def is_docker_installed():

    """Verifica si Docker está instalado en el sistema."""
    try:
        # Ejecuta 'docker --version' para verificar si Docker está instalado
        subprocess.run(["docker", "--version"], check=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        print("Docker está instalado.")
        return True

    except subprocess.CalledProcessError:
        print("Docker no está instalado.")
        return False



# Función para detectar el sistema operativo

def detect_os():

    """Detecta el sistema operativo en el que se ejecuta el script."""
    os_name = platform.system().lower()

    if os_name == "windows":
        return "windows"

    elif os_name == "darwin":
        return "macos"

    elif os_name == "linux":
        return "ubuntu"

    else:
        raise Exception("Sistema operativo no soportado.")



# Función para instalar Docker en Windows

def install_docker_windows():

    """Instala Docker en Windows."""
    print("Instalando Docker en Windows...")

    try:
        subprocess.run(["winget", "install", "Docker.DockerDesktop"], check=True)
        print("Docker instalado correctamente en Windows.")

    except subprocess.CalledProcessError:
        print("Error al intentar instalar Docker en Windows.")
        sys.exit(1)



# Función para instalar Docker en macOS

def install_docker_macos():

    """Instala Docker en macOS."""
    print("Instalando Docker en macOS...")

    try:
        subprocess.run(["brew", "install", "docker"], check=True)
        print("Docker instalado correctamente en macOS.")

    except subprocess.CalledProcessError:
        print("Error al intentar instalar Docker en macOS.")
        sys.exit(1)



# Función para instalar Docker en Ubuntu

def install_docker_ubuntu():

    """Instala Docker en Ubuntu."""
    print("Instalando Docker en Ubuntu...")

    try:
        subprocess.run(["sudo", "apt-get", "update"], check=True)
        subprocess.run(["sudo", "apt-get", "install", "-y", "docker.io"], check=True)
        subprocess.run(["sudo", "systemctl", "start", "docker"], check=True)
        subprocess.run(["sudo", "systemctl", "enable", "docker"], check=True)
        print("Docker instalado correctamente en Ubuntu.")

    except subprocess.CalledProcessError:
        print("Error al intentar instalar Docker en Ubuntu.")
        sys.exit(1)



# Función para crear un archivo docker-compose.yml

def create_docker_compose_file():

    """Crea el archivo docker-compose.yml con la configuración básica para Spark."""
    print("Creando archivo docker-compose.yml...")

    docker_compose_content = """
    version: "3"
    services:
      master:
        image: bitnami/spark:latest
        container_name: spark-master
        environment:
          - SPARK_MODE=master
        ports:
          - "8080:8080"
          - "7077:7077"
      worker:
        image: bitnami/spark:latest
        container_name: spark-worker
        environment:
          - SPARK_MODE=worker
          - SPARK_MASTER=spark://spark-master:7077
        depends_on:
          - master
    """

    try:
        with open("docker-compose.yml", "w") as f:
            f.write(docker_compose_content)
        print("Archivo docker-compose.yml creado correctamente.")

    except Exception as e:
        print(f"Error al crear el archivo docker-compose.yml: {e}")
        sys.exit(1)



# Función principal

def main():

    """Función principal que coordina la instalación de Docker y la creación de docker-compose.yml."""
    # Detecta el sistema operativo
    os_name = detect_os()
    

    # Verifica si Docker está instalado
    if not is_docker_installed():

        if os_name == "windows":
            install_docker_windows()

        elif os_name == "macos":
            install_docker_macos()

        elif os_name == "ubuntu":
            install_docker_ubuntu()

        else:
            print("Sistema operativo no soportado para la instalación de Docker.")
            sys.exit(1)

    
    # Verifica si Docker Compose está instalado
    if shutil.which("docker-compose") is None:

        print("Docker Compose no está instalado. Instalando...")

        try:
            subprocess.run(["sudo", "apt-get", "install", "-y", "docker-compose"], check=True)
            print("Docker Compose instalado correctamente.")

        except subprocess.CalledProcessError:
            print("Error al intentar instalar Docker Compose.")
            sys.exit(1)
    
    # Crea el archivo docker-compose.yml
    create_docker_compose_file()
    
    # Inicia el clúster de Docker usando docker-compose
    # print("Iniciando el clúster de Spark...")

    # try:
        # subprocess.run(["docker-compose", "up", "-d"], check=True)
        # print("Clúster de Spark iniciado correctamente.")

    # except subprocess.CalledProcessError:
        # print("Error al intentar iniciar el clúster con Docker Compose.")
        # sys.exit(1)



# Ejecuta la función principal
if __name__ == "__main__":
    main()
```
